```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dt=pd.read_csv('/content/Tasla_Stock_Updated_V2.csv')
```

```python
dt.head()
```

```python
dt.tail()
```

```python
sns.scatterplot(x='Open',y='Close',data=dt,hue='Volume')
```

```python
sns.scatterplot(x='High',y='Low',data=dt,hue='Volume')
```

```python
x=dt.drop(columns=['Close','Date'])
y=dt['Close']
```

```python
 from sklearn.linear_model import LinearRegression
 from sklearn.model_selection import train_test_split
```

```python
sns.lineplot(data=dt,x='Open',y='Close')
sns.lineplot(data=dt,x='High',y='Low')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_hat=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn import metrics
```

```python
mean_squared_error(y_test,y_hat)
```

```python
metrics.explained_variance_score(y_test, y_hat)
```

```python
plt.plot(y_hat[0:10])
plt.plot(np.array(y_test[0:10]))
plt.legend(["PREDICTED", "ACTUAL"])
```
